```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

```python
plotset = pd.read_csv('data/plot_df.csv')
```

```python
values = plotset.status_binary.value_counts().values
```

```python
values_sum = plotset.status_binary.value_counts().values.sum()
```

```python
fig = plt.figure(figsize=(8,8))
labels = ['good loans', 'bad loans']
pos = np.arange(len(labels))
percentages = [values[0]/values_sum*100, values[1]/values_sum*100]
colors = ['green']
colors.extend(['red']*(len(labels)-1))
bars = plt.bar(labels, values, color=colors)
i=0
for rect in bars:
        height = rect.get_height()
        plt.text(rect.get_x() + rect.get_width()/2.0, height - 180, '%d' % int(percentages[i]) + '%', ha='center', va='bottom',
                color = 'black', fontsize = 22, weight = 'semibold')
        i=i+1
plt.ylabel('Occurrence on sampled dataset', fontsize=18)
plt.xticks(pos, labels, alpha = 0.8, fontsize=20)
plt.savefig('graphics/loan_count.png')
```

```python
badset = plotset[plotset['status_binary']==0]
x = badset['amount_borrowed']
fig = plt.figure(figsize=(8,6))
plt.hist(x, color='grey')
plt.xlabel('Amount borrowed (bad loans only)', fontsize=20)
plt.ylabel('Frequence on sampled set', fontsize=18)
total = badset['amount_borrowed'].sum()
plt.title('Total: $' + '%d' % int(total) , fontsize=18)
plt.savefig('graphics/AmountDist_Bad.png')
```

```python
badset = plotset[plotset['status_binary']==1]
x = badset['amount_borrowed']
fig = plt.figure(figsize=(8,6))
plt.hist(x, color='grey')
plt.xlabel('Amount borrowed (good loans only)', fontsize=20)
plt.ylabel('Frequence on sampled set', fontsize=18)
total = badset['amount_borrowed'].sum()
plt.title('Total: $' + '%d' % int(total) , fontsize=18)
plt.savefig('graphics/AmountDist_Good.png')
```
