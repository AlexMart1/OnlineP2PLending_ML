Using a subsampled set to do some playground visualizations with the features deemed of more importance according to the Random Forest Classifier

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
sample_df = pd.read_csv('data/regraded_sample_df.csv')
X = sample_df.iloc[:,:-1]
y = sample_df.iloc[:,-1]
pca = PCA(n_components=2).fit_transform(X)
```

```python
plt.figure(dpi=80)
plt.scatter(pca[:,0],pca[:,1])
plt.title('Raw PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
```

```python
plt.figure(dpi=120)
#Scatter plot PC for only the Bad loans (y.values==0)
plt.scatter(pca[y.values==0,0], pca[y.values==0,1], alpha=0.5, label='Bad loans', s=2)
#Scatter plot PC for only the Good loans (y.values==1)
plt.scatter(pca[y.values==1,0], pca[y.values==1,1], alpha=0.5, label='Good loans', s=2)
plt.legend()
plt.title('Lending Data Set - First Two Principal Components')
plt.xlabel('PC1')
plt.ylabel('PC2')
```
